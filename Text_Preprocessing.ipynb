```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re
df = pd.read_csv('covid19_tweets.csv')
df.head()
```

Links Removal

```python
df["clean_text"] = df["text"].apply(lambda s: ' '.join(re.sub('http://\S+|https://\S+', " ", s).split()))
df['text'] == df['clean_text']
```

Punctuation Removal

```python
df['clean_text'] = df['clean_text'].apply(lambda s: ' '.join(re.sub("[.,!?:;-='...\"@#_]", " " , s).split()))
df[['text' , 'clean_text']]
```

Numbers Removal

```python
df['clean_text'].replace('\d+','', regex=True , inplace=True)
df[['text','clean_text']]
```

Emojis Removal

```python
df[['text','clean_text']].loc[12]
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii'))
df[['text','clean_text']].loc[12]
```

Stopwords Removal

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop = set(stopwords.words('english'))
print(stop)
```

```python
def remove(string):
    words = string.lower().split()
    clean_words = [word for word in words if word not in stop] 
    clean_string = " ".join(clean_words) 
    return clean_string
```

```python
df["clean_text"] = df["clean_text"].apply(lambda s: remove(s))
df[["text","clean_text"]].head()
```
